```python
def get_state_district_data(base_url, state_fips, type, fields, legislative_district, key, session):
    params = {'get': fields, 'for': f'state legislative district ({type} chamber):{legislative_district}', 'key': key, 'in': f'state:{state_fips}'}
    req = requests.Request('GET', base_url, params=params)
    prepared_req = session.prepare_request(req)
    #print(prepared_req.url)
    resp = session.get(base_url, params=params)
    resp = resp.json()
    return (float)(pd.DataFrame(resp).iloc[1, 0])
    #resp.pop(0)
    #return (int)(resp[0][0])
```

```python
def plot(df, x, y, title):
    plt.figure(figsize=(10, 6))
    plt.plot(df[x], df[y], marker='o', linestyle='-')
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.show()
```

```python
def get_stats(state_fips, chamberType, fields, legislative_district, key, session):
    stats = []
    stats.append(get_2010(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2011(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2012(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2013(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2014(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2015(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2016(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2017(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2018(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2019(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2020(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2021(state_fips, chamberType, fields, legislative_district, key, session))
    stats.append(get_2022(state_fips, chamberType, fields, legislative_district, key, session))
    return stats
```
